```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel("/content/varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# APP e APlicativo são as mesma coisa
# vamos renomar o APP para Aplicativo

vendas["idcanalvenda"] = vendas["idcanalvenda"].str.replace("APP", "Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
# Tirar o espaço dos Departamentos (Boa Prática)
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ", "_")
#vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(",", "")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
#Verificar os NULLs

vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna();
vendas_sem_nulos.info()
```

```python
# Uma premissa é que estada nullo deverá ser considerado o estado de MS

# vendas['estado'] = vendas['estado'].fillna("MS")
# ou
vendas['estado'].fillna("MS", inplace=True)
vendas.info()
```

```python
# Acertando o preço
# Substituir os nullos pela a média de todos os preços
media_preco = vendas['Preço'].mean()
vendas["Preço"].fillna(media_preco, inplace=True)
vendas.info()
```

```python
vendas[['Preço','Preço_com_frete']].describe()
```

```python
# Preço nao pode ser maior que o Frete
preco_errado = vendas.query("Preço > Preço_com_frete")
vendas_correta = vendas.query("Preço < Preço_com_frete")
```

```python
vendas_correta.describe()
```

```python
vendas_correta.head()
```

```python
vendas_correta.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP' ").describe()
```

```python
# Departamentos mais Vendidos
vendas_correta.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# Média de preço com frete por nome de departamento
# round -> Casas decimais
round(vendas_correta.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
# Quantidade de vendas por Mês
vendas_correta["mes"] = vendas_correta["Data"].dt.month
#vendas_correta.head()
vendas_correta.groupby("mes").idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
vendas_correta.head()
```

```python
cliente[["idade", "renda"]].describe()
```

```python
cliente.info()
```

```python
cliente = cliente.astype({"renda":"float"})
cliente.info()
```

```python
vendas_cliente = vendas_correta.merge(cliente, how="left", on="cliente_Log")
vendas_cliente.head()
```

```python
# Média de Renda para cada tipo de canal de venda
agg_idcanal_renda = vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index()
round(agg_idcanal_renda, 2)
```

```python
# Média de idade Clientes por Bandeira
agg_idade_bandeira = vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index()
round(agg_idade_bandeira, 2)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(7,4))
#agg_idade_bandeira
plt.bar(agg_idade_bandeira["bandeira"], agg_idade_bandeira["idade"], color="green")
plt.ylabel("Média de Idade")
plt.title("Média de idade de Bandeira", loc="center", fontsize=18, fontweight="bold", color="black" )
```

```python
fig = plt.figure(figsize=(7,4))
#agg_idade_bandeira

plt.bar(agg_idcanal_renda["idcanalvenda"], agg_idcanal_renda["renda"], color="green")
plt.ylabel("Média de renda")
plt.title("Média de Renda por Canal", loc="center", fontsize=18, fontweight="bold", color="black" )
```

```python
vendas_por_data = vendas_correta.groupby("Data").idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize=(20,4))
plt.xlabel("Dada da venda")
plt.ylabel("Quantidade de vendas")
x = vendas_por_data["Data"]
y = vendas_por_data["idcompra"]
plt.plot(x,y, color="green", linewidth="2")
```

```python
agg_dept_preco = round(vendas_correta.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(), 2)
agg_dept_preco
```

```python
fig = plt.figure(figsize=(20,7))
plt.xlabel("Departamento")
plt.ylabel("Preço com Frete")
plt.title("Preço médio por departamento", loc="center", fontsize=18, fontweight="bold", color="black")
x = agg_dept_preco["Nome_Departamento"]
y = agg_dept_preco["Preço_com_frete"]
plt.bar(x,y, color="green")
```
